```python
import pandas as pd
```

```python
import requests

url = "https://amazon-product-data6.p.rapidapi.com/product-by-text"

querystring = {"keyword":"SAMSUNG GALAXY","page":"1","country":"IN"}

headers = {
	"X-RapidAPI-Key": "28d4518340msh17e4d3d7c29290ap1df918jsn9a77d8f70884",
	"X-RapidAPI-Host": "amazon-product-data6.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
data=pd.json_normalize(response.json(),record_path=["data"])
```

```python
data.head()
```

```python
asin=data["asin"]
```

```python

for i in asin:
    url = "https://amazon-product-data6.p.rapidapi.com/product-review"

    querystring = {"asin":i,"page":"1","country":"IN","reviewer_type":"all_reviews","filter_by_star":"all_stars","sort_by":"helpful"}

    headers = {
    "X-RapidAPI-Key": "28d4518340msh17e4d3d7c29290ap1df918jsn9a77d8f70884",
    "X-RapidAPI-Host": "amazon-product-data6.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
data1=pd.json_normalize(response.json(),record_path=["data", "reviews"])
```
